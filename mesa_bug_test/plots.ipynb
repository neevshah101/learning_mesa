```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import scienceplots

plt.style.use('science')
```

```python
df = pd.read_table('LOGS1_no_evolve2/history.data', skiprows = 5, sep = '\s+')
```

```python
df.columns
```

```python
plt.figure(figsize = (6,4))
lw = 2
plt.plot(df['age'], df['star_1_mass'], label = 'Donor', lw = lw)
plt.plot(df['age'], df['star_2_mass'], label = 'Accretor', lw = lw)
plt.plot(df['age'], df['star_1_mass'] + df['star_2_mass'], label = 'Total', lw = lw)

plt.xscale('log')
plt.xlim(1.070e7,1.080e7)

plt.legend()
plt.xlabel('Age')
plt.ylabel('Mass')
plt.title('evolve_both_stars = .false.')
```

```python
plt.figure(figsize = (6,4))
plt.plot(df['age'], df['lg_mstar_dot_1'], label = 'Donor', lw = 2)
plt.plot(df['age'], df['lg_mstar_dot_2'], label = 'Accretor', lw = 2, linestyle = 'dashed')

plt.xscale('log')
plt.xlim(1.070e7,1.080e7)

plt.ylim(-9,-2)

plt.legend()
plt.xlabel('Age')
plt.ylabel('lg_mstar_dot')
plt.title('evolve_both_stars = .false.')
```
