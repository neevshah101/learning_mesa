```python
def plot_x_y(x,y,xlabel,ylabel,xscale = 'linear', yscale = 'linear', lw = 2, color = 'C0', label = False):
    plt.plot(x,y, color = color, lw = lw)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xscale(xscale)
    plt.yscale(yscale)

def plot_x_2y(x,y1,y2,xlabel,ylabel,xscale = 'linear', yscale = 'linear', lw = 2, label1 = 'y1', label2 = 'y2'):
    plt.plot(x,y1,lw = lw,label = label1)
    plt.plot(x,y2,lw = lw,label = label2)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xscale(xscale)
    plt.yscale(yscale)
    plt.legend()
```

```python
plot_x_y(dfB['star_mass'],10**dfB['log_R'],xlabel = 'Mass', ylabel = 'Radius')
plt.show()

plot_x_y(dfB['star_age']/1e3,dfB['lg_mstar_dot_1'],xlabel = 'Age (kyr)', ylabel = 'log M1 dot')
plt.ylim(-9,0)
plt.xlim(10670,10700)
plt.show()

plot_x_y(dfB['star_age']/1e3,dfB['period_days'],xlabel = 'Age (kyr)', ylabel = 'Period (days)', yscale = 'log')
plt.xlim(10670,10700)
plt.show()

plot_x_2y(dfB['star_age']/1e6, 10**(dfB['log_R']),dfB['rl_1'], xlabel = 'Age (Myr)', ylabel = 'Radius', label1 = 'Stellar Radius', label2 = 'Roche Radius')
plt.xlim(10,11)
plt.show()

plot_x_2y(dfB['star_age']/1e6, dfB['star_1_mass'],dfB['star_2_mass'], xlabel = 'Age (Myr)', ylabel = 'Mass', label1 = 'Donor Mass', label2 = 'Accretor Mass')
plt.xlim(10.6,10.8)
plt.show()

plot_x_2y(dfB['star_age']/1e6, dfB['total_mass_h1'],dfB['total_mass_he4'], xlabel = 'Age (Myr)', ylabel = 'Mass', label1 = 'H1 Mass', label2 = 'He4 Mass')
#plt.xlim(6,12)
plt.show()
```
