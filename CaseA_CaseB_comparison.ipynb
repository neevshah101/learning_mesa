```python
def plot_x_y(x,y,xlabel,ylabel,xscale = 'linear', yscale = 'linear', lw = 2, color = 'C0', label = False):
    plt.plot(x,y, color = color, lw = lw)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xscale(xscale)
    plt.yscale(yscale)

def plot_x_2y(x,y1,y2,xlabel,ylabel,xscale = 'linear', yscale = 'linear', lw = 2, label1 = 'y1', label2 = 'y2'):
    plt.plot(x,y1,lw = lw,label = label1)
    plt.plot(x,y2,lw = lw,label = label2)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xscale(xscale)
    plt.yscale(yscale)
    plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the figure
plt.figure(figsize=(10, 6))

# Scatter plot for ages < 3e6
mask1 = dfb['star_age'] < 3.1e6
sc1 = plt.scatter(dfb['log_Teff'][mask1], dfb['log_L'][mask1], c=dfb['star_age'][mask1], cmap='viridis', label='Ages < 3e6')

# Scatter plot for ages between 3e6 and 3.9e6
mask2 = (dfb['star_age'] >= 3.1e6) & (dfb['star_age'] < 3.9e6)
sc2 = plt.scatter(dfb['log_Teff'][mask2], dfb['log_L'][mask2], c=dfb['star_age'][mask2], cmap='plasma', label='Ages 3e6 to 3.9e6')

# Invert the x-axis
plt.gca().invert_xaxis()

# Create colorbars
cbar1 = plt.colorbar(sc1, ax=plt.gca(), orientation='vertical')
cbar1.set_label('Star Age (up to 3e6)')
cbar1.set_ticks([dfb['star_age'][mask1].min(), 1e6, 2e6, 3.1e6])
cbar1.set_ticklabels([f'{dfb["star_age"][mask1].min():.1e}', '1e6', '2e6', '3.1e6'])

# Create a finer second colorbar
cbar2 = plt.colorbar(sc2, ax=plt.gca(), orientation='vertical')
cbar2.set_label('Star Age (3.1e6 to 3.9e6)')

# Set more ticks for the finer colorbar
fine_ticks = np.linspace(3.1e6, 3.9e6, num=10)  # 10 evenly spaced ticks
cbar2.set_ticks(fine_ticks)
cbar2.set_ticklabels([f'{tick:.1e}' for tick in fine_ticks])  # Format the tick labels

# Label axes
plt.xlabel('Log Effective Temperature')
plt.ylabel('Log Luminosity')

# Show the plot
plt.legend()
plt.show()
```
